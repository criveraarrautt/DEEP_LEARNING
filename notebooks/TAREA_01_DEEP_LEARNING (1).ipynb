```python
from PIL import Image, ImageDraw, ImageFont

# Crear una nueva imagen en blanco
image = Image.new('RGB', (500, 500), (255, 255, 255))
draw = ImageDraw.Draw(image)

# Añadir un círculo como parte del diseño minimalista
draw.ellipse((150, 150, 350, 350), outline='black', width=5)

# Usar la fuente predeterminada
font = ImageFont.load_default()
text = "NLP"
text_width, text_height = draw.textsize(text, font=font)
draw.text(((500 - text_width) / 2, (500 - text_height) / 2), text, fill="black", font=font)

# Guardar la imagen
image.save('logo_nlp.png')

```

##Leer el Notebook de los roles en la ciencia de datos:##

| **Rol**                         | **Descripción**                                                                                                                                               |
|---------------------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **Arquitecto de datos**         | Define las políticas, los procedimientos, los modelos y las tecnologías que se utilizarán para recopilar, organizar, almacenar y acceder a la información.    |
| **Ingeniero de datos**          | Crea canalizaciones de datos para reunir información de diferentes sistemas de origen.                                                                         |
| **Analista de negocio**         | Procesa, interpreta y documenta procesos comerciales, productos, servicios y software a través del análisis de datos para ayudar a tomar decisiones.          |
| **Analista de datos**           | Responde preguntas analizando los datos e interpretando los resultados.                                                                                        |
| **Investigador de IA**          | Explora y propone nuevos sistemas para resolver problemas que luego son utilizados por otros roles en escenarios del mundo real.                              |
| **Científico de datos**         | Se enfoca en resolver problemas mediante el desarrollo de modelos basados en IA más que dar respuestas a preguntas.                                            |
| **Ingeniero de ML**             | Crea y mantiene sistemas de inteligencia artificial para automatizar modelos predictivos.                                                                      |
| **Ingeniero de visualización**  | Diseña soluciones de informes de datos que simplifiquen la comprensión de los datos y sus conocimientos.                                                      |
| **Ingeniero de MLOps**          | Brinda a los científicos de datos y otros roles acceso a las herramientas e infraestructura especializadas necesarias a lo largo del ciclo de vida de los datos.|
